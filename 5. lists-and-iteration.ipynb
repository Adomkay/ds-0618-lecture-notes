```python
import csv
albums = []
with open('./data.csv') as foo:
    reader = csv.DictReader(foo)
    for row in reader:
        albums.append(row)
```

```python

def clean_albums(albums):
    for album in albums:
        album['number'] = int(album['number'])
        album['year'] = int(album['year'])
    return albums
    
clean_albums(albums)
```

```python
artists = [album['artist'] for album in albums]
histogram = dict.fromkeys(artists, 0)
# allocate space for a pile of each artist

for artist in artists:
    histogram[artist] += 1
```

```python

def find_top_artists(histogram):
    max_albums = max(list(histogram.values()))
    album_tuples = list(histogram.items())
    top_artists = []
#     ('the who', 3)
    for album_tuple in album_tuples:
        if album_tuple[1] == max_albums:
            top_artists.append(album_tuple)
    return top_artists
```
